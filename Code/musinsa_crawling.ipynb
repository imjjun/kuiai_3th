#Crawling

```python
import os
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
texts = []
tags = []
img_urls = []

hdr = {'User-Agent': 'Mozilla/5.0'} #봇 차단 방지를 위한 헤더 추가

for num in range(13596, 31122+1): #2020년 1월 ~ 현재(2023년 1월 16일)까지의 무신사 코디숍 사진

    url = "https://www.musinsa.com/app/styles/views/"+str(num)
    req = Request(url,headers=hdr) 
    page = urlopen(req)
    soup = BeautifulSoup(page)
    
    try:
        text = soup.select_one('#style_info > div.styling_tag > p').text #style text 
        tag = soup.select_one('#style_info > div.styling_tag > div').text.split('\n') #tag
        img_url = soup.select_one('img')['src'] #첫번째 이미지 url
        
        texts.append(text)
        tags.append(tag)
        img_urls.append(img_url)
        
    except AttributeError: #유효하지 않은 콘텐츠인 경우 pass
        pass
    
    if num%1000==0:
        print("complete", num)
```

```python
print(len(texts),len(tags), len(img_urls))
```

#Dataframe

```python
df = pd.DataFrame(list(zip(img_urls, tags, texts)), columns = ['img_url', 'tag', 'text']) #데이터 프레임 변환
```

```python
df.to_csv("musinsa.csv",index=False) #df csv파일로 저장
```
